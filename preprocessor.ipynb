```python
def percent_capital(title):
    number_capitals = sum(1 for c in str(title) if c.isupper())
    number_overall = len(str(title))
    return number_capitals / number_overall
```

```python
def title_length(title):
    return len(str(title))
```

```python
def question_mark(title):
    if "?" in str(title): 
        return True
    else:
        return False
```

```python
def exclamation_mark(title):
    if "!" in str(title): 
        return True
    else:
        return False
```

```python
def fully_capital(title):
    tokens = str(title).split()
    for token in tokens:
        if(token.isupper()):
            return True
    return False
```

```python
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
def get_sentiment(title):
    return sia.polarity_scores(str(title))["compound"]
```

```python
def number_fully_capital(title):
    count = 0
    tokens = str(title).split()
    for token in tokens:
        if(token.isupper()):
            count+=1
    return count
```

# SCRUBBING

```python
# for scraped data
df = df[df['video_id'] != 'video_id']
```

```python
df = df.drop(columns=['video_id', 'trending_date', 'channelTitle', 'categoryId', 'publishedAt', 'tags', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'description', 'channelId','comment_count','likes','dislikes'])
df
```

# FEATURE CREATION

```python
m = df['title'].apply(isEnglish)
df = df[m]
df
```

```python
df['percent_capital'] = df['title'].apply(percent_capital)
df['title_length'] = df['title'].apply(title_length)
df['contains_question_mark'] = df['title'].apply(question_mark)
df['contains_exclamation_mark'] = df['title'].apply(exclamation_mark)
df['VADER'] = df['title'].apply(get_sentiment)
df['has_full_cap_word'] = df['title'].apply(fully_capital)
df['full_cap_count'] = df['title'].apply(number_fully_capital)
```

```python
def tanh_normalizer(value, column_name):
    return 0.5 * (np.tanh(0.01 * ( value - np.mean(df[column_name])) / np.std(df[column_name])) + 1)
```

# TANH NORMALIZATION

```python
df['title_length'] = df['title_length'].apply(lambda value : tanh_normalizer(value, 'title_length'))
df['VADER'] = df['VADER'].apply(lambda value : tanh_normalizer(value, 'VADER'))
df['full_cap_count'] = df['full_cap_count'].apply(lambda value : tanh_normalizer(value, 'full_cap_count'))
```
